```python
## import necessary libraries
import requests
import pandas as pd

# Function to fetch data from Alpha Vantage API
def get_stock_data(symbol, api_key):
    base_url = "https://www.alphavantage.co/query"
    params = {
        "function": "TIME_SERIES_DAILY",
        "symbol": symbol,
        "apikey": api_key,
        "outputsize": "full"
    }
    response = requests.get(base_url, params=params)
    if response.status_code == 200:
        data = response.json()
        # Convert data to pandas DataFrame
        df = pd.DataFrame(data["Time Series (Daily)"]).T
        # Rename columns
        df.columns = ['open', 'high', 'low', 'close', 'volume']
        # Convert index to datetime
        df.index = pd.to_datetime(df.index)
        # Sort DataFrame by index in descending order (most recent first)
        df = df.sort_index(ascending=False)
        return df
    else:
        print(f"Failed to fetch data for {symbol}")
        return None

# Grab the data
api_key = "YBE5UJA188L4G65L"  # Replace with your Alpha Vantage API key
symbols = ["MSFT", "AAPL", "NVDA", "AVGO", "TSM", "ASML", "ORCL", "CRM", "AMD", "SAP", "ADBE", "CSCO", "ACN", "QCOM", "INTU", "AMAT", "TXN", "IBM", "NOW", "UBER", "INTC", "MU", "LRCX", "ARM", "SONY", "ADI"]
for symbol in symbols:
    stock_data = get_stock_data(symbol, api_key)
    if stock_data is not None:
        print(f"Stock data for {symbol}:")
        print(stock_data.head())
        print()  # Add empty line between each stock's data

# Create a DataFrame for a company table
company_data = pd.DataFrame({
    "ticker": symbols,
    "company_name": ["Microsoft", "Apple", "Nvidia", "Broadcom", "Taiwan Semiconductor", "ASML", "Oracle", "Salesforce", "AMD", "SAP", "Adobe", "Cisco", "Accenture", "Qualcomm", "Intuit", "Applied Materials", "Texas Instruments", "IBM", "ServiceNow", "Uber Technologies", "Intel", "Micron", "Lam Research", "ARM", "Sony", "Analog Devices"]
})

```
